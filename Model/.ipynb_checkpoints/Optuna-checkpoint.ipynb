# Optuna

```python
%%time

import optuna
from xgboost import XGBClassifier

from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import log_loss, make_scorer
```

```python
%%time

cv       = StratifiedKFold(n_splits=4,shuffle=True, random_state=1)
```

```python
# Define Objective Function.

def OjbectiveFunc(trial):
    
    # set Parameters    
    param = {
        
        "verbosity": 0,
        "n_estimators": trial.suggest_int("n_estimators", 20, 70),
        # defines booster, gblinear for linear functions.
        "booster": trial.suggest_categorical("booster", ["gbtree", "gblinear", "dart"]),
        # L2 regularization weight.
        "lambda": trial.suggest_float("lambda", 1e-8, 1.0, log=True),
        # L1 regularization weight.
        "alpha": trial.suggest_float("alpha", 1e-8, 1.0, log=True),
        # sampling ratio for training data.
        "subsample": trial.suggest_float("subsample", 0.2, 1.0),
        # sampling according to each tree.
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.2, 1.0),
        # maximum depth of the tree, signifies complexity of the tree.
        "max_depth" : trial.suggest_int("max_depth", 3, 9),
        # minimum child weight, larger the term more conservative the tree.
        "min_child_weight" : trial.suggest_int("min_child_weight", 2, 10),
        "eta" : trial.suggest_float("eta", 1e-8, 1.0, log=True),
        # defines how selective algorithm is.
        "gamma" : trial.suggest_float("gamma", 1e-8, 1.0, log=True),
        "grow_policy" : trial.suggest_categorical("grow_policy", ["depthwise", "lossguide"]),


    }

    
    # Form Model
    clf = XGBClassifier( **param )
    
    # Evaluate Score
    SCORE = cross_val_score( clf, X_train, y_train, scoring='neg_log_loss', cv=cv, n_jobs=None, verbose=0 ).mean()
    
    return SCORE
```

```python
## Form Study
study = optuna.create_study(direction="maximize")
```

```python
study.optimize(OjbectiveFunc, n_trials=20, timeout=None, n_jobs=2, show_progress_bar=True)
```

```python
study.best_params
```

```python
best_model = XGBClassifier(use_label_encoder=False, n_jobs=4, **study.best_params)
```

```python
best_model.fit(X_train, y_train)
```

```python
y_pred = pd.DataFrame(best_model.predict_proba(X_test), columns=sub.columns)
```

```python
y_pred.isna().sum()
```

```python
y_pred.to_csv("../SUBS/OopTunaSup10.csv", index=False)
```
